```python
import mne

# Replace with your file path
raw = mne.io.read_raw_edf('/content/physionet.org/files/chbmit/1.0.0/chb01/chb01_01.edf', preload=True)
raw.set_eeg_reference('average')  # Set common average reference
```

```python
# Dictionary for filtered signals by frequency band
filter_bank = {
    'delta': raw.copy().filter(0.5, 4, fir_design='firwin'),
    'theta': raw.copy().filter(4, 8, fir_design='firwin'),
    'alpha': raw.copy().filter(8, 13, fir_design='firwin'),
    'beta':  raw.copy().filter(13, 30, fir_design='firwin'),
    'gamma': raw.copy().filter(30, 80, fir_design='firwin')
}
```

```python
# Step 1: Remove powerline noise (notch filter)
raw.notch_filter(freqs=[50, 100])  # Or 60Hz for US

# Step 2: Baseline correction
raw.apply_function(lambda x: x - x.mean(), picks='eeg')

# Step 3: Interpolation (if any bad channels are marked)
# raw.info['bads'] = ['EEG 053']  # Example
# raw.interpolate_bads()

# Step 4: Resample if needed
raw_resampled = raw.copy().resample(256)

# Step 5: Epoching (sliding window, 4s with 40% overlap)
events = mne.make_fixed_length_events(raw_resampled, duration=4, overlap=1.6)
epochs = mne.Epochs(raw_resampled, events, tmin=0, tmax=4, baseline=None, preload=True)
```

```python
import numpy as np
from scipy.stats import kurtosis, skew
from scipy.signal import welch

def extract_features(epoch_data, sfreq):
    feats = []
    for ch_data in epoch_data:
        # Time-domain features
        mean = np.mean(ch_data)
        var = np.var(ch_data)
        skewness = skew(ch_data)
        kurt = kurtosis(ch_data)

        # Frequency domain (PSD)
        freqs, psd = welch(ch_data, sfreq)
        psd_mean = np.mean(psd)
        psd_var = np.var(psd)

        feats.extend([mean, var, skewness, kurt, psd_mean, psd_var])
    return np.array(feats)

# Apply feature extraction
X = np.array([extract_features(epoch, raw.info['sfreq']) for epoch in epochs.get_data()])
```

```python
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler

# Normalize first
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dummy target (you need actual labels)
y = np.random.randint(0, 2, size=X.shape[0])

# Apply LASSO
lasso = LassoCV(cv=5).fit(X_scaled, y)
X_selected = X_scaled[:, lasso.coef_ != 0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne

# Use one of the uploaded EDF files (user needs to ensure it's uploaded)
edf_file_path = "/content/physionet.org/files/chbmit/1.0.0/chb01/chb01_01.edf"
# Load the EEG data
raw = mne.io.read_raw_edf(edf_file_path, preload=True)

# Apply basic preprocessing: band-pass filter (0.5–40 Hz), remove DC offset, resample
raw.filter(0.5, 40., fir_design='firwin')
raw.apply_function(lambda x: x - np.mean(x), picks='eeg')
raw.resample(256)

# Select 10 seconds of EEG from first 2 channels
start, stop = raw.time_as_index([0, 10])
data, times = raw[:, start:stop]

# Plot preprocessed EEG signals
plt.figure(figsize=(12, 5))
plt.plot(times, data[0], label=raw.ch_names[0])
plt.plot(times, data[1], label=raw.ch_names[1])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (uV)')
plt.title('Preprocessed EEG Signal (First 10 seconds)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
